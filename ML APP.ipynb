```python
import pandas as pd
df =  pd.read_csv("raw_data2.csv", error_bad_lines=False)
def dfreplace(df, *args, **kwargs):
    s = pd.Series(df.values.flatten())
    s = s.str.replace(*args, **kwargs)
    return pd.DataFrame(s.values.reshape(df.shape), df.index, df.columns)

df = dfreplace(df, ',', '')
```
