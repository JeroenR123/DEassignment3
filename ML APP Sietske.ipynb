```python
import pandas as pd
df =  pd.read_csv("raw_data2.csv", error_bad_lines=False)
def dfreplace(df, *args, **kwargs):
    s = pd.Series(df.values.flatten())
    s = s.str.replace(*args, **kwargs)
    return pd.DataFrame(s.values.reshape(df.shape), df.index, df.columns)

df = dfreplace(df, ',', '')
```

```python
df
```

```python
import re
import numpy as np
for i in df.columns:
    df[i] = df[i].astype(str)
    df[i][df[i].apply(lambda i: True if re.search('^\s*$', str(i)) else False)]=np.NaN
```

```python
df['YYYY'] = df['YYYYMMDD,'].str.slice(0,4) #create a variable for years
df['MM'] = df['YYYYMMDD,'].str.slice(4,6)#create a variable for months
df['DD'] = df['YYYYMMDD,'].str.slice(6,8)
for i in df.columns:
        df[i] =df[i].astype(float, errors= 'ignore') 
df = df.drop('YYYYMMDD,', axis=1)
```

```python
train_df = train_df.drop(columns = ['STN,','EV24', 'NG,', 'TN,', 'TNH,', 'TX,', 'TXH,', 'T10N,', 'T10NH,'])
train_df = train_df.dropna()
```

## Machine Learning Model

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LinearRegression
from sklearn import metrics
import math

# get x and y
x_train, y_train = train_df.drop('TG,', axis=1), train_df['TG,']
# split the data for initial testing
X_train, X_test, Y_train, Y_test = train_test_split(x_train, y_train, test_size=0.2,random_state=1)
X_train
```

```python
model = LinearRegression()
model.fit(X_train, Y_train)

print(model.coef_)
```

```python
predictions = model.predict(X_test)
```

```python
predictions
```

```python
df_test_results = pd.DataFrame({'Prediction':predictions, 'Real': Y_test})
df_test_results
```

```python
MSE = metrics.mean_squared_error(df_test_results['Real'],df_test_results['Prediction'])
RMSE = math.sqrt(MSE)
MAXE = max((df_test_results['Real']- df_test_results['Prediction'])**2)
MAE = metrics.mean_absolute_error(df_test_results['Real'],df_test_results['Prediction'])

pd.Series([MSE,RMSE,MAXE,MAE], index = ['MSE','RMSE','MAXE','MAE'])
```

## Model using Univeriate selection

```python
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.feature_selection import chi2

kBestSelector = SelectKBest(f_classif, k=6) 
kBestSelector.fit(X_train, Y_train)

ixkBest = kBestSelector.get_support() 
X_train_kBest = pd.DataFrame(kBestSelector.transform(X_train), columns = X_train.columns.values[ixkBest])
X_test_kBest = pd.DataFrame(kBestSelector.transform(X_test), columns = X_train.columns.values[ixkBest])
X_train_kBest.head()
```

```python
model2 = LinearRegression()
model2.fit(X_train_kBest, Y_train)

print(model2.coef_)
```

```python
predictions = model2.predict(X_test_kBest)
df_test_results = pd.DataFrame({'Prediction':predictions, 'Real': Y_test})
df_test_results
```

```python
MSE = metrics.mean_squared_error(df_test_results['Real'],df_test_results['Prediction'])
RMSE = math.sqrt(MSE)
MAXE = max((df_test_results['Real']- df_test_results['Prediction'])**2)
MAE = metrics.mean_absolute_error(df_test_results['Real'],df_test_results['Prediction'])

pd.Series([MSE,RMSE,MAXE,MAE], index = ['MSE','RMSE','MAXE','MAE'])
```
